```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data = {
    'area': [500, 750, 1000, 1200, 1500, 1800, 2000, 2200],
    'price': [800, 1000, 1300, 1500, 1800, 2100, 2300, 2500]
}
pf=pd.DataFrame(data)
pf
```

```python
%matplotlib inline
plt.xlabel('area(sq ft)')
plt.ylabel('price(usd)')
plt.scatter(pf.area,pf.price)
```

```python
reg=linear_model.LinearRegression()
reg.fit(pf[['area']],pf.price)
```

```python
reg.predict(np.array([4000]).reshape(-1, 1))
```

```python
%matplotlib inline
plt.scatter(pf[['area']],pf.price)
plt.plot(pf.area,reg.predict(pf[['area']]))
```
